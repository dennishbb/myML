```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Create figure and axis
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_xlim(0, 10)
ax.set_ylim(0, 7)
ax.axis('off')

# Define phase positions (shifted upward for separation)
phases = [
    ("Preparation\n(Q1 2025)\n(Jan - Mar)", 1, 6),
    ("Migration Execution\n(Q2Q3 2025)\n(Apr - Sep)", 4, 6),
    ("Post-Migration &\nOptimization (Q4 2025)\n(Oct - Dec)", 7, 6)
]

# Define tasks within each phase (leveled positions)
tasks = [
    (1, 4.0, "Assess Compatibility"),
    (1, 3.5, "Plan Infra Changes"),

    (4, 4.0, "Deploy on E1/E2 with RHEL 8"),
    (4, 3.5, "Test and Validation"),
    (4, 3.0, "Application & Workload Testing"),
    (4, 2.5, "Production Deployment"),

    (7, 4.0, "Monitor System Performance"),
    (7, 3.5, "Apply Security & OS Updates"),
    (7, 3.0, "Retire RHEL 7.9 Instances")
]

# Draw phase rectangles separately at the top
for phase, x, y in phases:
    ax.add_patch(mpatches.FancyBboxPatch((x-0.6, y-0.3), 3.2, 0.8,
                                         boxstyle="round,pad=0.1",
                                         edgecolor='black', facecolor='#a7c7e7'))
    ax.text(x+1, y, phase, ha='center', va='center', fontsize=10, fontweight='bold', color='black', wrap=True)

# Draw task rectangles separately below phases
for x, y, task in tasks:
    ax.add_patch(mpatches.FancyBboxPatch((x-0.3, y-0.15), 2.5, 0.5,
                                         boxstyle="round,pad=0.1",
                                         edgecolor='black', facecolor='#f4d03f'))
    ax.text(x+1, y+0.05, task, ha='center', va='center', fontsize=8, color='black')

# Draw arrows from phases to first task in each phase
for i in range(len(phases)):
    ax.arrow(phases[i][1] + 1, phases[i][2] - 0.4, 0, -1, head_width=0.2, head_length=0.2, fc='black', ec='black')

# Display the diagram with improved spacing
plt.title("Upgrade Plan: Cloud Pak for Data System 1.0.9.0 with RHEL 8.x", fontsize=14, fontweight='bold', color='black')
plt.show()

```
